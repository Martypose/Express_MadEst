```python
# binarizar imagen de escala de grises
#  hacer mascaras, normalmente parte blanca la que nos interesa

import cv2 as cv
import numpy as np

from matplotlib import pyplot as plt
import pylab # this allows you to control figure size 
pylab.rcParams['figure.figsize'] = (10.0, 8.0) # this controls figure size in the notebook

img = cv.imread("tablas.png")
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)






plt.imshow(gray, cmap='gray')

```

```python
blur = cv.GaussianBlur(gray, (13, 13), 0)
edges = cv.Canny(blur, 46, 21, apertureSize=3)

plt.imshow(edges, cmap='gray')
```

```python
minLineLength=10
lines = cv.HoughLinesP(image=edges,rho=1,theta=np.pi/180, threshold=100,lines=np.array([]), minLineLength=minLineLength,maxLineGap=15)

a,b,c = lines.shape
for i in range(a):
    cv.line(img, (lines[i][0][0], lines[i][0][1]), (lines[i][0][2], lines[i][0][3]), (0, 0, 255), 3, cv.LINE_AA)
```
